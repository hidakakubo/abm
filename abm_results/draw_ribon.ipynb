```python
# ライブラリのインポート
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import random
from tqdm import tqdm
import statistics
import os
import itertools
import math

%matplotlib inline
```

```python
def main():
    # パラメータの設定
    seed = 1000
    x_name = "step"
    y_name_1 = "population_size"
    y_name_2 = "male_proportion"

    # データの読み込み
    data_list = []
    for i in range(1, seed+1):
        data = pd.read_csv(f"results/result{i}.csv")           # ディレクトリ構造に注意！
        data_list.append(data)

    data_all = pd.concat(data_list, ignore_index=True)

    # 図1
    fig = plt.figure(figsize=(10, 7))
    ax = sns.lineplot(x=x_name, y=y_name_1, data=data_all)
    print(type(ax))
    ax.set_xlabel("generation")
    ax.set_ylabel("population size")
    fig.savefig('tmp1.png', dpi=600)

    # 図2
    fig = plt.figure(figsize=(10, 7))
    ax = sns.lineplot(x=x_name, y=y_name_2, data=data_all)
    print(type(ax))
    ax.set_xlabel("generation")
    ax.set_ylabel("male proportion")
    fig.savefig('tmp2.png', dpi=600)


if __name__ == '__main__':
    main()
```
