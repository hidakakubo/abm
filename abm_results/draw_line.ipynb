```python
# ライブラリのインポート
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
from tqdm import tqdm
import statistics
import os
import itertools
import math

%matplotlib inline
```

```python
def main():
    # パラメータの設定
    iteration = 50  # 世代数

    # データの読み込み
    df = pd.read_csv('results/result.csv')
    P_size = df['population_size']
    male_prop = df['male_proportion']

    # 時間(横軸)
    t = list(range(iteration + 1))

    # 可視化
    # 図1
    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.plot(t, P_size)
    ax.set_xlabel('Generations[year]')
    ax.set_ylabel('population size')
    ax.grid(True)
    fig.savefig('tmp1.png', dpi=600)

    # 図2
    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.plot(t, male_prop)
    ax.set_xlabel('Generations[year]')
    ax.set_ylabel('male prop')
    ax.grid(True)
    fig.savefig('tmp2.png', dpi=600)

    fig.show()


if __name__ == '__main__':
    main()
```
